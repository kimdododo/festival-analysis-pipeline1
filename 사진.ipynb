```python
# -*- coding: utf-8 -*-
# 강서구 펌프장 내/외수위 예측 - 멀티모델 비교 (화면 출력 전용)
import os, re, platform, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# ========= 0) 공통 설정: 저장 대신 바로 출력 =========
SHOW_PLOTS = True   # <- 화면 출력 on/off (False로 바꾸면 저장)
OUTDIR = "figs"     # 저장 모드일 때만 사용

def render(fig=None, path=None, close=True):
    """SHOW_PLOTS=True면 plt.show(), False면 파일 저장"""
    if fig is None:
        fig = plt.gcf()
    if SHOW_PLOTS:
        plt.show()
    else:
        os.makedirs(os.path.dirname(path) if path else OUTDIR, exist_ok=True)
        if not path:
            path = os.path.join(OUTDIR, "plot.png")
        fig.savefig(path, dpi=220, bbox_inches="tight")
    if close:
        plt.close(fig)

# ========= 1) 한글 폰트 =========
def patch_korean_font():
    sysname = platform.system().lower()
    fonts = (["Malgun Gothic","맑은 고딕"] if "windows" in sysname
             else ["AppleGothic"] if "darwin" in sysname
             else ["NanumGothic","DejaVu Sans"])
    for f in fonts:
        try:
            matplotlib.rcParams["font.family"] = f
            _ = plt.figure(); plt.title("폰트테스트"); plt.close()
            print(f"[INFO] 폰트 적용: {f}"); break
        except Exception: continue
    matplotlib.rcParams["axes.unicode_minus"] = False
patch_korean_font()

# ========= 2) 파일 경로 & tol =========
PUMP_FILE  = "1.csv"
HYDRO_FILE = "hydro_raw_2017_2025.csv"
RAIN_FILE  = "rain_2017_2024_merged.csv"
SEWER_FILE = "sewer_2017_2024_merged.csv"
RAIN_TOL, HYDRO_TOL, SEWER_TOL = "1H", "1H", "30min"

# ========= 3) 데이터 로드/정규화 =========
def load_pump(p=PUMP_FILE):
    df = pd.read_csv(p)
    if "timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
    elif {"일자","시각"} <= set(df.columns):
        ts_a = pd.to_datetime(df["일자"], errors="coerce")
        ts_b = pd.to_datetime(df["시각"], errors="coerce")
        df["timestamp"] = ts_a.combine_first(ts_b)
    else:
        raise KeyError("펌프 데이터에 'timestamp' 또는 ['일자','시각'] 필요")
    return df.sort_values("timestamp").reset_index(drop=True)

def load_rain(p=RAIN_FILE):
    r = pd.read_csv(p)
    r["timestamp"] = pd.to_datetime(r["일시"], errors="coerce") if "일시" in r.columns else pd.to_datetime(r["timestamp"], errors="coerce")
    if "강수량(mm)" in r.columns: r = r.rename(columns={"강수량(mm)":"rain_mm"})
    r["rain_mm"] = pd.to_numeric(r["rain_mm"], errors="coerce")
    return r[["timestamp","rain_mm"]].dropna(subset=["timestamp"]).sort_values("timestamp")

def load_hydro(p=HYDRO_FILE):
    h = pd.read_csv(p)
    if "관측 일시" not in h.columns: raise KeyError("수문 데이터에 '관측 일시' 필요")
    date_col = "관측 일시"
    hour_cols = [c for c in h.columns if re.fullmatch(r"\s*\d{1,2}시\s*", str(c))]
    long = h.melt(id_vars=[date_col], value_vars=hour_cols, var_name="hour_k", value_name="hydro_value")
    long[date_col] = pd.to_datetime(long[date_col], errors="coerce").dt.floor("D")
    long["hour_num"] = long["hour_k"].astype(str).str.extract(r"(\d{1,2})").astype(int)
    long["timestamp"] = long[date_col] + pd.to_timedelta((long["hour_num"]-1)%24, unit="h")
    long["hydro_value"] = pd.to_numeric(long["hydro_value"], errors="coerce")
    return (long.dropna(subset=["timestamp"])
                .groupby("timestamp", as_index=False)["hydro_value"].mean()
                .sort_values("timestamp"))

def load_sewer(p=SEWER_FILE):
    s = pd.read_csv(p)
    if not {"timestamp","water_level"} <= set(s.columns): raise KeyError("하수 데이터에 'timestamp','water_level' 필요")
    s["timestamp"] = pd.to_datetime(s["timestamp"], errors="coerce")
    s["water_level"] = pd.to_numeric(s["water_level"], errors="coerce")
    return s.groupby("timestamp", as_index=False)["water_level"].mean().sort_values("timestamp")

def merge_asof_(L, R, tol):
    if R is None or len(R)==0: return L
    return pd.merge_asof(L.sort_values("timestamp"),
                         R.sort_values("timestamp"),
                         on="timestamp", direction="backward",
                         tolerance=pd.Timedelta(tol))

pump, rain, hydro, sewer = load_pump(), load_rain(), load_hydro(), load_sewer()
df = merge_asof_(pump, rain, RAIN_TOL)
df = merge_asof_(df,   hydro, HYDRO_TOL)
df = merge_asof_(df,   sewer, SEWER_TOL)

# 결측 보정
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df[num_cols] = df[num_cols].ffill().bfill()
for c in num_cols: df[c] = df[c].fillna(df[c].median())

# ========= 4) 시계열 파생 (펌프장명 그룹) =========
def add_time_features(_df, group_col="펌프장명",
                      base=("rain_mm","hydro_value","water_level","내수위","외수위")):
    d = _df.copy()
    if group_col in d.columns:
        d = d.sort_values([group_col,"timestamp"]); g = d.groupby(group_col, group_keys=False)
        def roll(gser, w): return gser.rolling(w).mean().reset_index(level=0, drop=True)
        for col in base:
            if col in d.columns and pd.api.types.is_numeric_dtype(d[col]):
                d[f"{col}_lag1"]  = g[col].shift(1)
                d[f"{col}_lag2"]  = g[col].shift(2)
                d[f"{col}_diff1"] = g[col].diff()
                d[f"{col}_rm3"]   = roll(g[col], 3)
                d[f"{col}_rm6"]   = roll(g[col], 6)
    else:
        d = d.sort_values("timestamp")
        for col in base:
            if col in d.columns and pd.api.types.is_numeric_dtype(d[col]):
                d[f"{col}_lag1"]  = d[col].shift(1)
                d[f"{col}_lag2"]  = d[col].shift(2)
                d[f"{col}_diff1"] = d[col].diff()
                d[f"{col}_rm3"]   = d[col].rolling(3).mean()
                d[f"{col}_rm6"]   = d[col].rolling(6).mean()
    # 보정
    for c in d.select_dtypes(include=[np.number]).columns: d[c] = d[c].fillna(d[c].median())
    return d

df = add_time_features(df)

# ========= 5) 모델 정의 =========
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor

models = {
    "Linear": Pipeline([("sc", StandardScaler()), ("m", LinearRegression())]),
    "Ridge":  Pipeline([("sc", StandardScaler()), ("m", Ridge(alpha=1.0))]),
    "Lasso":  Pipeline([("sc", StandardScaler()), ("m", Lasso(alpha=1e-3, max_iter=10000))]),
    "RandomForest": RandomForestRegressor(n_estimators=300, random_state=42, n_jobs=-1),
    "GBDT": GradientBoostingRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42),
    "HistGB": HistGradientBoostingRegressor(max_iter=500, learning_rate=0.05, max_depth=6, random_state=42),
}
try:
    from xgboost import XGBRegressor
    models["XGBoost"] = XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6,
                                     subsample=0.8, colsample_bytree=0.8,
                                     random_state=42, n_jobs=-1, tree_method="hist")
except Exception as e:
    print("[INFO] XGBoost 스킵:", e)
try:
    from lightgbm import LGBMRegressor
    models["LightGBM"] = LGBMRegressor(n_estimators=500, learning_rate=0.05, max_depth=-1,
                                       subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
except Exception as e:
    print("[INFO] LightGBM 스킵:", e)
try:
    from catboost import CatBoostRegressor
    models["CatBoost"] = CatBoostRegressor(iterations=500, learning_rate=0.05, depth=6,
                                           random_state=42, loss_function="RMSE", verbose=False)
except Exception as e:
    print("[INFO] CatBoost 스킵:", e)

from sklearn.model_selection import GroupKFold, KFold
from sklearn.metrics import mean_squared_error, r2_score

# ========= 6) 타깃별 학습/OOF/시각화 (화면 출력) =========
TARGETS = [t for t in ["내수위","외수위"] if t in df.columns]
DROP = {"timestamp","일자","시각","펌프장 주소","위도","경도","데이터기준일자"}

for TARGET in TARGETS:
    print(f"\n==================== [타겟: {TARGET}] ====================")
    feats = [c for c in df.columns if c not in DROP|{TARGET} and df[c].dtype!="O"]
    X = df[feats].copy(); y = df[TARGET].copy()
    mask = X.notna().all(axis=1) & y.notna(); X, y = X[mask], y[mask]

    if "펌프장명" in df.columns:
        splits = lambda: GroupKFold(n_splits=5).split(X, y, groups=df.loc[X.index, "펌프장명"])
    else:
        splits = lambda: KFold(n_splits=5, shuffle=True, random_state=42).split(X)

    results = []
    for name, base_model in models.items():
        # OOF 예측
        oof = np.zeros(len(y))
        for tr_idx, va_idx in splits():
            try:
                model = base_model.__class__(**base_model.get_params())
            except Exception:
                model = base_model
            model.fit(X.iloc[tr_idx], y.iloc[tr_idx])
            oof[va_idx] = model.predict(X.iloc[va_idx])

        rmse = float(np.sqrt(mean_squared_error(y, oof)))
        r2v  = float(r2_score(y, oof))
        results.append((name, rmse, r2v))
        print(f"[{name:10s}] RMSE={rmse:.4f}  R²={r2v:.5f}")

        # --- 회귀선 (OOF) ---
        fig = plt.figure(figsize=(5.2,4.2))
        plt.scatter(y, oof, s=10, alpha=0.45)
        lo, hi = float(min(y.min(), oof.min())), float(max(y.max(), oof.max()))
        plt.plot([lo,hi],[lo,hi],"r--",lw=1)
        k = np.polyfit(y, oof, 1)
        xx = np.linspace(lo,hi,100); plt.plot(xx, k[0]*xx+k[1], "g-", lw=1.2)
        plt.title(f"{TARGET} - {name} (OOF)\nRMSE={rmse:,.2f}  R²={r2v:.5f}")
        plt.xlabel("실제값"); plt.ylabel("예측값")
        render(fig)  # ← 저장 대신 바로 출력

        # --- 변수 중요도 (가능 모델만, FULL 적합) ---
        try:
            mfit = base_model.__class__(**base_model.get_params()); mfit.fit(X, y)
            if hasattr(mfit, "feature_importances_"):
                imp = (pd.DataFrame({"feature":feats, "importance":mfit.feature_importances_})
                         .sort_values("importance", ascending=False).head(20))
                fig = plt.figure(figsize=(8,6))
                plt.barh(imp["feature"][::-1], imp["importance"][::-1])
                plt.title(f"{TARGET} - {name} 중요 변수 Top20 (FULL)")
                render(fig)
        except Exception as e:
            print(f"[INFO] 중요도 스킵({name}): {e}")

        # --- SHAP (가능한 경우만) ---
        try:
            import shap
            ns = min(1000, len(X)); Xs = X.sample(ns, random_state=42)
            mfit = base_model.__class__(**base_model.get_params()); mfit.fit(X, y)
            explainer = shap.Explainer(mfit, Xs); sh = explainer(Xs)

            shap.summary_plot(sh, Xs, plot_type="bar", show=SHOW_PLOTS, max_display=20)
            if not SHOW_PLOTS: render(plt.gcf())
            shap.summary_plot(sh, Xs, plot_type="dot", show=SHOW_PLOTS, max_display=20)
            if not SHOW_PLOTS: render(plt.gcf())
        except Exception as e:
            print(f"[INFO] SHAP 스킵({name}): {e}")

    # 성능표 출력
    res = pd.DataFrame(results, columns=["Model","RMSE","R2"]).sort_values("RMSE")
    print("\n[성능 비교]\n", res.to_string(index=False))

print("\n[DONE] 모든 결과를 화면에 출력했습니다.")
```
